# 📊 E-Commerce Sales Analysis using Python

## 📌 Project Overview
This project analyzes e-commerce sales data to identify trends and business insights.

## 🎯 Objectives
- Analyze sales performance
- Identify top categories and products
- Understand profit trends

## 🛠 Tools Used
- Python
- Pandas
- NumPy
- Matplotlib
- Seaborn


## 📂 Dataset Information

The dataset includes:
- Order Date
- Category
- Product Name
- Region
- Segment
- Sales and Profit


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'order_date': ['2020-01-01','2020-02-01','2020-03-01','2020-04-01',
                   '2020-05-01','2020-06-01'],
    'category': ['Furniture','Technology','Office Supplies','Furniture',
                 'Technology','Office Supplies'],
    'product_name': ['Chair','Laptop','Pen','Table','Mobile','Notebook'],
    'region': ['East','West','South','East','North','West'],
    'segment': ['Consumer','Corporate','Home Office','Consumer','Corporate',
                'Home Office'],
    'total_sales': [200,500,100,300,700,150],
    'profit': [20,80,10,50,120,30]
}
df = pd.DataFrame(data)
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().drop_duplicates()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['Year'] = df['order_date'].dt.year
df['Month'] = df['order_date'].dt.month
```

```python
df['total_sales'].sum()
```

```python
df.groupby('category')['total_sales'].sum()
```

```python
df.groupby('region')['total_sales'].sum()
```

```python
df.groupby('product_name')['total_sales'].sum().sort_values(ascending=False)
```

```python
sns.barplot(x='category',y='total_sales',data=df)
plt.show()
```

```python
df.groupby('Month')['total_sales'].sum().plot()
plt.show()
```

```python
sns.barplot(x='segment',y='profit',data=df)
plt.show()
```

## 📊 Key Insights

- Technology category has the highest sales compared to others.
- Sales are increasing over the months, showing a positive trend.
- Corporate segment generates more profit than other segments.
- Some regions are performing weaker and need improvement.
- High sales do not always mean high profit .


```python
df.groupby('category')[['total_sales','profit']].sum()
```

```python
df['profit_ratio'] = df['profit'] / df['total_sales']
```

```python
df.groupby('category')['profit_ratio'].mean()
```

```python
df.groupby('product_name')['profit'].sum().sort_values(ascending=False).head(3)
```

```python
pivot = df.pivot_table(values='total_sales',index='region',columns='category')
sns.heatmap(pivot , annot=True)
plt.show()
```

## 📊 Advanced Insights

- Profit ratio shows business efficiency across categories.
- Some categories generate high sales but low profit.
- Heatmap highlights region-wise performance differences.
- Top products contribute most to overall profit.
